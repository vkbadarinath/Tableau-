Q1. What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/
Cloud in terms of accessibility and data security?

A)*The key difference is that dashboards published on Tableau Public are openly accessible to anyone on the internet with no data security controls, while dashboards on Tableau Server/Cloud are restricted to authorized users within an organization, offering enterprise‑grade security, compliance, and controlled accessibility.*

#Accessibility

- Tableau Public

   - Dashboards are publicly visible to anyone with the link or by searching online.

  - No authentication or user restrictions — ideal for sharing with a broad audience (e.g., blogs, portfolios, academic projects).

   -C annot connect to private or sensitive live data sources; data is extracted and published openly.

- Tableau Server / Tableau Cloud

   - Dashboards are accessible only to licensed users or groups defined by administrators.

   - Supports role‑based access control (viewer, editor, admin).

   - Enables secure sharing within an organization, with integration to identity providers (Active Directory, SSO).

#Data Security

- Tableau Public

   - No data security — once published, the workbook and data are stored on Tableau’s public servers.

   - Anyone can download and reuse the workbook unless permissions are restricted to “view only.”

   - Not suitable for confidential, regulated, or proprietary datasets.

- Tableau Server / Tableau Cloud

   - Enterprise‑grade security with encryption, authentication, and compliance (ISO, SOC, GDPR).

   - Organizations retain control over data governance:

   - Tableau Server → hosted on‑premises or private cloud, full control over infrastructure and compliance.

   - Tableau Cloud → fully managed SaaS by Tableau, with strong operational security and automatic updates.

   - Supports granular permissions, audit trails, and integration with corporate IT policies.

Q2. When publishing a Tableau dashboard, what happens to a live connection compared to an extract?
How does this affect data refresh and performance?

A)
#Live Connection

- **Behavior when published:** The dashboard continues to query the underlying database in real time. Each time a user opens or interacts with the dashboard, Tableau sends queries directly to the source.

- **Data Refresh:** Always up‑to‑date, since it pulls the latest data from the source. No need to schedule refreshes.

- **Performance Impact:**

   - Dependent on the speed and load of the database/network.

   - If the source is large or slow, dashboards may lag.

   - Heavy user traffic can put strain on the database.

#Extract

- **Behavior when published:** Tableau creates a snapshot of the data and stores it in a highly optimized .hyper file.

- **Data Refresh:** Not automatic. You must schedule refreshes (on Tableau Server/Cloud) or manually refresh extracts to keep data current.

- **Performance Impact:**

   - Faster queries, since Tableau reads from the in‑memory extract rather than hitting the live database.

   - Reduces load on the source system.

   - Best for large datasets or when the source is slow/unreliable.

#Trade‑off Summary

- Live Connection = Always current, but performance depends on source system.

- Extract = Faster performance and reduced database load, but data may be stale unless refreshes are scheduled.

Q3 In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users
can view, interact with, or edit a published dashboard?

A)*In Tableau Server or Tableau Cloud, the feature that allows administrators to control which users can view, interact with, or edit a published dashboard is role‑based permissions (also called user roles and permissions management).*

- Roles: Users are assigned roles such as Viewer, Explorer, Creator, or Administrator.

- Permissions: Each role defines what actions a user can perform:

   - Viewer → Can only see and interact with dashboards.

   - Explorer → Can interact, filter, and create new views but not publish.

   - Creator → Can build, publish, and edit dashboards.

   - Administrator → Full control over users, groups, and content.

- Granular Control: Permissions can be set at the project, workbook, or dashboard level, allowing fine‑grained access management.

- Integration: Works with enterprise identity systems (Active Directory, SSO) for secure authentication.

Q4 If you want users to view a dashboard but not download the underlying data, which permission setting
would you adjust in Tableau Server?

A)
#How Permissions Work

- Download Full Data → Allows users to export the entire underlying dataset.

- Download Summary Data → Allows users to export aggregated or summarized data.

- Download Workbook/Save a Copy → Lets users download the entire workbook (.twb/.twbx).

By denying these permissions while still granting View and Interact, users can explore the dashboard online but cannot extract or export the raw data behind it.

#Best Practice

- Grant View and Filter/Interact permissions.

- Deny Download Full Data, Download Summary Data, and Download Workbook.

- This ensures dashboards remain interactive but the sensitive dataset stays protected.

Q6 List two best practices you should follow when preparing a dashboard for public sharing to ensure data
security and privacy.

A)
#1. Remove or Mask Sensitive Data

- Strip out confidential fields (e.g., customer names, IDs, financial details, health records).

- Use aggregated measures instead of raw transactional data.

- Apply data masking or anonymization techniques if you need to showcase patterns without exposing identities.

#2. Limit Download and Reuse Options

- Disable Download Full Data / Summary Data / Workbook permissions before publishing.

- Share only the visual insights, not the underlying dataset.

- This prevents unauthorized users from extracting raw data for misuse.

Q7 After publishing a dashboard, users report a “Data not found” error.
 What is the most common cause related to data source settings?

A)*The most common cause of a “Data not found” error after publishing a Tableau dashboard is that the data source connection settings were not configured for server access.*

#Key Issue

- When you build a dashboard locally (in Tableau Desktop), the connection may point to a local file path (Excel, CSV, or text file) or a database connection using your personal credentials.

- After publishing, Tableau Server/Cloud cannot access that local file or may not have the right credentials to connect to the database.

- As a result, the dashboard loads but the data source is missing, triggering the “Data not found” error.

#Common Causes

- Local file paths (Excel/CSV stored on your computer, not on a shared server).

- Database credentials not embedded during publishing.

- Extracts not published along with the workbook.

- Permissions on the data source not granted to the Server/Cloud environment.

Q8 When embedding a Tableau dashboard into a secure internal portal, what two additional security
measures should be implemented to ensure only authorized users can access the embedded content?

A)*When embedding a Tableau dashboard into a secure internal portal, two additional security measures you should implement are:*

#1. Authentication Integration

- Ensure the portal uses Single Sign‑On (SSO) or integrates with your organization’s identity provider (e.g., Active Directory, Okta, Azure AD).

- This way, only authenticated employees can access the embedded dashboard.

- Prevents anonymous or external users from bypassing Tableau’s security by simply accessing the embed link.

#2. Role‑Based Permissions in Tableau

- Configure permissions on Tableau Server/Cloud so only authorized groups (e.g., “Finance Team,” “Executives”) can view the dashboard.

- Deny download/export permissions if sensitive data is involved.

- Even if the embed code is shared, unauthorized users won’t be able to see the content without proper Tableau access rights.
