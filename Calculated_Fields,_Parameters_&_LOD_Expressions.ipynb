#Theoretical Questions :

Q1. What is a Calculated Field and why is it important in business dashboards?

A)*A Calculated Field is a custom formula you create inside a dashboard tool (like Power BI, Tableau, or Excel) to derive new metrics from existing data. It’s important because it transforms raw numbers into meaningful business insights, enabling KPIs, trend analysis, and decision‑making that raw data alone cannot provide.*

#What is a Calculated Field?

- Definition: A calculated field is a new column or metric created by applying formulas to existing fields in your dataset.

- Example: If you have Sales and Profit, you can create a calculated field called Profit Margin = Profit ÷ Sales.

- Key Point: It doesn’t alter the source data; instead, it dynamically generates new values inside the dashboard environment.

#Importance of Calculated Fields in Dashboards

1. Enable Custom KPIs

    - Businesses rarely rely only on raw metrics like Sales or Expenses.

   - Calculated fields let you build Profit Margin, ROI, Conversion Rate, Growth %, etc.—the exact KPIs leadership wants to track.

2. Support Strategic Decisions

    - Dashboards are meant for decision‑makers, not just analysts.

    - Calculated fields provide summarized, business‑oriented metrics that highlight performance, risks, and opportunities.

3. Provide Flexibility Without Changing Source Data

    - You don’t need to alter the database schema.

    - Calculated fields allow analysts to experiment with formulas directly in the dashboard tool, making insights faster and more adaptable.

4. Enable Trend & Comparative Analysis

    - Growth rates, year‑to‑date totals, rolling averages, and variance calculations are all powered by calculated fields.

    - This helps leaders see patterns over time rather than just static numbers.

5. Improve Communication Across Teams

    - Standardized calculated fields ensure that finance, sales, and operations teams all look at the same definition of metrics.

    - This avoids confusion and keeps reporting consistent.

6. Drive Real-Time Monitoring

    - Since dashboards update automatically, calculated fields provide live KPIs (e.g., Gross Margin %, Customer Churn Rate) that reflect the current state of the business.

Q2.  Why can’t parameters change visuals directly in Tableau?

A)
- Parameters are just single-value inputs.  
They hold a value (like a number, string, or date) but do not interact with the data source by themselves.
Example: A parameter might store “Region = East” or “Discount = 10%”.

- Parameters don’t filter or aggregate data directly.  
Unlike filters, parameters don’t know how to slice the dataset. They simply provide a value that you can use in a calculated field or reference line.

- Calculated Fields are the bridge.  
To make a parameter affect a visualization, you must connect it through a calculated field.

$Ex:-{IF [Region] = [Region Parameter] THEN [Sales] END}$

- Design philosophy in Tableau:  
Tableau separates inputs (parameters) from logic (calculated fields) and outputs (visuals). This modular design ensures flexibility and avoids accidental changes to visuals without explicit logic.

Q3. When should EXCLUDE LOD be used?

A)
- You want a less granular aggregation than the view shows.

  - Example: Your view shows Sales by Region and Product, but you want to calculate Sales by Region only, ignoring Product.

  - $Formula:{EXCLUDE [Product] : SUM([Sales])}$

- You need to compare detail vs. summary.

  - Example: Show each product’s sales compared to the overall region’s sales.

  - Product sales = SUM([Sales])

  - Region sales (ignoring product) = {EXCLUDE [Product] : SUM([Sales])}

  - Then you can compute % of Region.

- You want to simplify calculations without removing dimensions from the view.

  - Sometimes you need to keep a dimension for display (like Product names) but not let it affect the calculation. EXCLUDE lets you do that.

*Use EXCLUDE LOD when you want to aggregate at a higher level than the current visualization shows, without removing dimensions from the view itself. It’s especially useful for comparisons between detailed and summarized values (like product vs. region totals).*

Q4. Difference between FIXED and INCLUDE LOD expressions?

A)
Use FIXED when you want a calculation that stays constant no matter how the view changes.

Use INCLUDE when you want to temporarily add more detail into the calculation while still summarizing at the current view level.

#FIXED LOD
- Definition: FIXED explicitly sets the level of detail for the calculation, regardless of what dimensions are in the view.

- Behavior: It ignores the visualization’s context and computes at the dimension(s) you specify.

- Use Case: When you need a consistent aggregation (like Region totals or Customer averages) that shouldn’t change as you add/remove dimensions in the visualization.

- $Ex:- {FIXED [Region] : SUM([Sales])}$

#INCLUDE LOD
- Definition: INCLUDE adds extra granularity to the calculation beyond what’s currently in the view.

- Behavior: It respects the visualization’s context but drills down further by including additional dimensions.

- Use Case: When you want to add detail temporarily (like product-level averages or customer-level counts) while still summarizing at a higher level in the visualization.

- $Ex:-{INCLUDE [Product] : AVG([Sales])}$
