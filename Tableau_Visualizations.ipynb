Q9. Chart Selection Justification

Explain why different charts were used for different business questions in this analysis.

A)1. Profitability Investigation (Scatter Plot)
Why Scatter Plot?

Scatter plots are ideal for showing the relationship between two continuous measures (Sales vs Profit).

Each dot represents an order, making it easy to spot loss‑making orders (negative profit) and high‑impact transactions.

Business Value:

Finance can quickly identify risky orders and investigate why high‑sales transactions sometimes lead to losses.

2. Business Growth Evaluation (Line Chart with Running Total)
Why Line Chart?

Line charts are best for time‑based trends.

Using a running total of Sales shows cumulative growth, making it clear whether the business is expanding consistently.

Business Value:

Senior management can assess overall growth trajectory and compare performance across months or years.

3. Category vs Region Performance (Bar Chart / Heatmap)
Why Bar Chart/Heatmap?

Bar charts allow easy comparison of Sales across Regions, broken down by Category.

Heatmaps provide a quick visual of strong vs weak areas using color intensity.

Business Value:

Operations can identify which regions are strong for each category and where improvement is needed.

4. Product Prioritization (Bubble Chart / Scatter Plot with Size Encoding)

- Why Bubble Chart?

  - Bubble charts extend scatter plots by adding a third dimension (Quantity as bubble size).

  - This enables a multi‑dimensional analysis: Sales, Profit, and Quantity together.

- Business Value:

  - Product teams can prioritize items that deliver maximum impact (high sales, high profit, large quantities) and reconsider those that sell well but hurt margins.
