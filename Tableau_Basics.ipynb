Q1. Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis.

Also mention any two industries where Tableau is commonly used and explain one use case.

A) *Tableau is a modern data visualization and analytics platform that helps organizations turn raw data into interactive, easy‑to‑understand insights. **1. Healthcare**

**2. Retail & E‑Commerce**

- Retailers use Tableau to analyze sales trends, customer behavior, and supply chain performance.

- Use Case: An e‑commerce company uses Tableau to visualize product sales by region, identify top‑performing categories, and optimize inventory.

Q8. What is the difference between a Worksheet and a Dashboard in Tableau?

Why is a dashboard more useful for decision-makers?

A)
#Between Worksheet

1. A worksheet is a single view that contains one chart or visualization (e.g., bar chart, line chart, map).

2. It is built by dragging dimensions and measures onto shelves (Rows, Columns, Filters, Marks).

3. Each worksheet focuses on analyzing one aspect of the data at a time.

4. Worksheets are mainly used by analysts or developers to explore data in detail.

5. Example: A worksheet showing Sales by Category or Profit by Sub‑Category.

#Dashboard in Tableau
1. A dashboard is a collection of multiple worksheets and other objects (like text, images, filters, and interactive controls) arranged together.

2. It provides a consolidated view of different KPIs and metrics in one place.

3. Dashboards allow interactivity — filters, highlight actions, drill‑downs can be applied across multiple charts simultaneously.

4. They are designed for decision‑makers, who need a big‑picture view rather than isolated charts.

5. Example: A dashboard combining Sales Trend over Time, Profit by Region, and Top Products by Sales.

#🔹 Why Dashboards Are More Useful for Decision‑Makers

- Holistic View: Dashboards bring together multiple KPIs and charts in one place, so leaders don’t need to switch between worksheets.

- Quick Insights: They allow executives to monitor performance at a glance and spot trends or issues immediately.

- Interactivity: Filters, drill‑downs, and highlight actions let decision‑makers explore data dynamically without building new charts.

- Storytelling: Dashboards present data in a structured, visual narrative that supports strategic decisions.

Q9.  What is Tableau Prep, and why is it important in real-world data projects?

What is Tableau Prep, and why is it important in real-world data projects?

Then explain how you would:

Remove null values

Remove duplicate records

Change incorrect data types (Answer in steps, no tool access required.)

A)
#🔹 What is Tableau Prep?

- Tableau Prep is a data preparation tool from Tableau that helps clean, shape, and combine data before analysis.

- It provides a visual, step‑by‑step workflow for tasks like removing nulls, fixing data types, merging tables, and reshaping datasets.

- It is designed to be user‑friendly, so analysts can prepare data without complex coding.

#🔹 Importance in Real‑World Data Projects

- Data Quality: Ensures the data used in dashboards is accurate and reliable.

- Efficiency: Automates repetitive cleaning tasks, saving time.

- Integration: Prepares data from multiple sources (Excel, databases, cloud apps) into one clean dataset.

- Auditability: Each step is documented in the Prep workflow, making it easy to trace transformations.

- Decision Support: Clean data leads to trustworthy dashboards, which drive better business decisions.

#🔹 Steps to Clean Data in Tableau Prep

**1. Remove Null Values**

- Open the dataset in Tableau Prep.

- Identify fields with null values (Prep highlights them).

- Step: Right‑click the field → choose Filter Values → Exclude Nulls.

- Result: All rows with nulls in that field are removed.

**2. Remove Duplicate Records**

- Tableau Prep automatically detects duplicates when you group or aggregate.

- Step: Select the field(s) that define uniqueness (e.g., Customer ID, Order ID).

- Use the Remove Duplicates option or apply Aggregate → Group by unique fields.

- Result: Only unique records remain.

**3. Change Incorrect Data Types**

- Sometimes fields are misclassified (e.g., a numeric field stored as text).

- Step: In the Profile Pane, click the data type icon (ABC for text, # for number, calendar for date).

- Select the correct type (e.g., change from String → Number, or String → Date).

- Result: Field is corrected and can be used properly in analysis.

Q10.  A retail manager wants answers to the following:

Which category generates the highest profit?

Which region performs poorly?

How does sales change over time?

a) Mention which charts you would use
b) Identify one key insight that management could derive
c) Explain how this helps in business decision-making

A)
#a) Charts to Use

**1. Which category generates the highest profit?  **

- → Use a Bar Chart (Category on X‑axis, SUM(Profit) on Y‑axis).

- Easy to compare profit across categories.

**2. Which region performs poorly?**

→ Use a Map Chart or Bar Chart by Region (Region on X‑axis, SUM(Profit) or SUM(Sales) on Y‑axis).

- Map highlights geographical differences; bar chart gives clear numeric comparison.

**3. How does sales change over time?**

→ Use a Line Chart (Order Date on X‑axis, SUM(Sales) on Y‑axis).

- Shows trends, seasonality, and growth/decline patterns.

#b) One Key Insight for Management

- Example: Technology category generates the highest profit, but the West region underperforms compared to others. Sales show strong peaks in November and December, indicating seasonal demand.

#c) How This Helps in Business Decision‑Making

1. Category Profitability: Management can focus on expanding high‑profit categories and reconsider discounts or costs in low‑profit ones.

2. Regional Performance: Identifies weak regions (e.g., West) where targeted marketing, better distribution, or pricing strategies are needed.

3. Sales Trend: Seasonal peaks guide inventory planning, staffing, and promotional campaigns to maximize revenue during high‑demand months.
