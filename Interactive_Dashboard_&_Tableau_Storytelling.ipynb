Q1. Explain the difference between a tableauDashboard and a Tableau Story Point. Give two healthcare-releated use cases.

A)*A Tableau Dashboard shows multiple charts together for interactive analysis, while a Tableau Story Point presents those dashboards or worksheets sequentially to guide an audience through a narrative. In healthcare, dashboards are ideal for real-time monitoring, while story points are better for step-by-step presentations of patient journeys or outcomes.*

**Key Differences Between Tableau Dashboard and Story Point**

#Dashboard

Definition: A single screen combining multiple worksheets, charts, and filters.

Purpose: Provides a consolidated, interactive view of KPIs and metrics.

Interactivity: High – filters, highlight actions, drill‑downs across charts.

Audience use: Decision‑makers monitoring performance in real time.

Best For: Operational monitoring, executive dashboards, daily reporting.

#Story Point

Definition: A sequence of dashboards/worksheets arranged like slides.

Purpose: Guides the audience through a narrative or analysis step by step.

Interactivity: Limited – focus is on storytelling flow rather than exploration.

Audience Use: Presenters explaining insights or processes in a structured order.

Best For: Presentations, lectures, or explaining patient journeys/outcomes.

**Healthcare Use Cases**

1. Hospital Performance Monitoring (Dashboard)
- Scenario: A hospital administrator needs to track bed occupancy, patient admissions, staff allocation, and treatment costs in real time.

- Dashboard Benefit: Multiple charts (admissions trend, occupancy rate, staff workload) can be viewed together with filters for department or region.

- Outcome: Enables quick operational decisions, e.g., reallocating staff to departments with high patient load.

2. Patient Treatment Journey Presentation (Story Point)

- Scenario: A medical research team wants to present the progression of cancer treatment outcomes to stakeholders.

- Story Point Benefit: Sequentially shows patient demographics, treatment phases, survival rates, and final outcomes. Each slide builds on the previous one, guiding the audience through the narrative.

- Outcome: Provides a clear, structured explanation of how treatment impacts patients over time, useful for conferences or board meetings.

Q2. what are actions in tableau? Explain how filter Action can help hospital management analyze patient data.

A)*In Tableau, Actions are interactive features that let users control how different sheets, dashboards, or data elements respond to clicks, selections, or hovers. They create dynamic links between views, making dashboards more engaging and exploratory.*

#Types of actions include:

- Filter Actions → filter data in one view based on a selection in another.

- Highlight Actions → emphasize related data points across views.

- URL Actions → open external web pages or resources.

- Go to Sheet Actions → navigate between dashboards or worksheets.

#How Filter Action Helps in Hospital Management

A Filter Action allows hospital administrators to drill down into patient data interactively.

Example Workflow:

1. A dashboard shows overall patient admissions by department.

2. When the administrator clicks on Cardiology, a filter action automatically updates other charts to show:

  - Patient demographics (age, gender).

  - Admission reasons (heart disease, arrhythmia, etc.).

  - Average length of stay and treatment costs.

3. This interactive filtering helps management quickly analyze specific patient groups without manually applying filters.

Healthcare Benefit:

  - Enables real-time insights into patient flow.

  - Helps identify resource allocation needs (e.g., more beds or staff in cardiology).

  - Supports data-driven decisions for improving patient care and hospital efficiency.
